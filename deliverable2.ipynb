```python
import re
from nltk.corpus import stopwords
import nltk

# Download the stopwords
nltk.download('stopwords')
import nltk
nltk.download('vader_lexicon')
```

```python
# Define stopwords and cleaning function
additional_stopwords = [
    'thank', 'calling', 'united', 'airlines', 'customer', 'service', 'name', 
    'help', 'today', 'welcome', 'appreciate', 'business', 'flying', 'hi', 
    'yeah', 'okay', 'alright', 'ugh', 'sigh', 'pause', 'typing', 'noises', 
    'really', 'hopefully', 'um', 'hmm', 'understand', 'look', 'check', 
    'see', 'take', 'make', 'get', 'give', 'switch', 'hear', 'think', 
    'frustrating', 'ridiculous', 'unfortunately', 'problem', 'hoping', 
    'ideal', 'able'
]
extra_stopwords = ['calling', 'typing', 'okay', 'thanks', 'youre', 'yeah', 'really', 'work', 'day', 'need']
total_stopwords = additional_stopwords + extra_stopwords
stop_words = set(stopwords.words('english'))
stop_words.update(total_stopwords)

# Function to clean transcript
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()  # Convert to lowercase
    text = ' '.join([word for word in text.split() if word not in stop_words])  # Remove stopwords
    return text

# Apply the cleaning function to transcripts
final_master_df['cleaned_transcript'] = final_master_df['call_transcript'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Vectorize the cleaned transcripts
vectorizer = CountVectorizer(stop_words='english', max_features=20)
X = vectorizer.fit_transform(final_master_df['cleaned_transcript'])

# Get the top keywords per call reason
final_master_df['primary_reason'] = final_master_df["primary_call_reason"]
call_reason_groups = final_master_df.groupby('primary_reason')

# Analyze top keywords for each call reason
for reason, group in call_reason_groups:
    print(f"Top keywords for call reason {reason}:")
    top_keywords = vectorizer.fit_transform(group['cleaned_transcript']).toarray().sum(axis=0)
    keywords = vectorizer.get_feature_names_out()
    top_keywords_dict = dict(zip(keywords, top_keywords))
    sorted_keywords = sorted(top_keywords_dict.items(), key=lambda x: x[1], reverse=True)
    print(sorted_keywords[0:10])  # Print top 10 keywords
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Separate data by primary reason
baggage_transcripts = final_master_df[final_master_df['primary_reason'] == 'baggage']['cleaned_transcript']
booking_transcripts = final_master_df[final_master_df['primary_reason'] == 'booking']['cleaned_transcript']

# Combine into a single dataset for TF-IDF calculation
transcripts = pd.concat([baggage_transcripts, booking_transcripts])

# Use TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=20)  # Top 20 words
tfidf_matrix = tfidf.fit_transform(transcripts)

# Get feature names (keywords)
tfidf_feature_names = tfidf.get_feature_names_out()

# Create a DataFrame for better visualization
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_feature_names)
```

```python
# Use TF-IDF Vectorizer with bigrams (two-word phrases)
tfidf_bigram = TfidfVectorizer(ngram_range=(2, 2), max_features=20)
tfidf_bigram_matrix = tfidf_bigram.fit_transform(transcripts)

# Get bigram feature names
tfidf_bigram_feature_names = tfidf_bigram.get_feature_names_out()

# Create a DataFrame for bigrams
tfidf_bigram_df = pd.DataFrame(tfidf_bigram_matrix.toarray(), columns=tfidf_bigram_feature_names)
```

```python
# Example: Using TF-IDF and K-Means Clustering
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Convert transcripts to TF-IDF matrix
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(final_master_df['cleaned_transcript'])

# Perform K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=42)
final_master_df['cluster'] = kmeans.fit_predict(tfidf_matrix)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to generate word cloud
def plot_wordcloud(cluster_data, cluster_number):
    text = ' '.join(cluster_data['cleaned_transcript'])
    wordcloud = WordCloud(width=800, height=400, max_words=100).generate(text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud for Cluster {cluster_number}")
    plt.show()

# Generate word clouds for each cluster
for cluster in final_master_df['cluster'].unique():
    cluster_data = final_master_df[final_master_df['cluster'] == cluster]
    plot_wordcloud(cluster_data, cluster)
```

```python
# Sentiment analysis using a library like VADER
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Calculate sentiment score for each transcript
final_master_df['sentiment'] = final_master_df['cleaned_transcript'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Filter by sentiment score
low_emotion_calls = final_master_df[final_master_df['sentiment'] >= 0.9959]  # Example threshold for low sentiment
```

```python

# Plot the distribution of sentiment scores
plt.figure(figsize=(10,6))
sns.histplot(final_master_df['sentiment'], bins=100, kde=True)
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.show()
```

```python
# Group by primary call reason and count the occurrences
call_reason_counts = final_master_df.groupby('primary_call_reason')['call_id'].count().reset_index()

# Sort by count to see the most common reasons
call_reason_counts.columns = ['primary_call_reason', 'count']
call_reason_counts = call_reason_counts.sort_values(by='count', ascending=False)

# Display the top call reasons
print(call_reason_counts.head(10))  # Adjust the number to see more or fewer reasons
```

```python
# Assume sentiment analysis has already been done and added to the DataFrame
sentiment_analysis = final_master_df.groupby('primary_call_reason')['sentiment'].mean().reset_index()

# Sort to see which call reasons have the lowest (most negative) sentiment
sentiment_analysis.columns = ['primary_call_reason', 'average_sentiment']
sentiment_analysis = sentiment_analysis.sort_values(by='average_sentiment', ascending=True)

# Display reasons with the most negative sentiment
print(sentiment_analysis.head(10))
```

```python
# Analyze the average call duration for each primary call reason
call_duration_analysis = final_master_df.groupby('primary_call_reason')['handle_time'].mean().reset_index()

# Sort by duration to identify the call reasons with the longest duration
call_duration_analysis.columns = ['primary_call_reason', 'average_duration']
call_duration_analysis = call_duration_analysis.sort_values(by='average_duration', ascending=False)

# Display the reasons with longest call duration
print(call_duration_analysis.head(10))
```
